```python
# Replace negative ages with median
customers["age"] = customers["age"].apply(lambda x: None if x < 0 else x)
customers["age"].fillna(customers["age"].median(), inplace=True)
customers
```

```python
# convert datatype
customers["age"]=customers["age"].round().astype(int)
customers
```

```python
# convert dates
customers["join_date"] = pd.to_datetime(customers["join_date"], errors="coerce")
customers
```

```python
# convert dates
sales["sale_date"] = pd.to_datetime(sales["sale_date"], errors="coerce")
sales
```

```python
products
```

```python
#Create a new column in the sales data to calculate total_sale_amount (quantity sold × unit price).

sales_merged = sales.merge(products[["product_id", "price"]], on="product_id", how="left")

# Create the new column: total_sale_amount = quantity_sold * price
sales_merged["total_sale_amount"] = sales_merged["quantity_sold"] * sales_merged["price"]
sales_merged["total_sale_amount"] = sales_merged["total_sale_amount"].round()
print(sales_merged)
```

```python
# merge all tables
merged_df = sales.merge(customers, on="customer_id", how="left").merge(products, on="product_id", how="left")
merged_df
```

```python
# a) Total sales by category
sales_by_category = merged_df.groupby("category")["sale_amount"].sum().reset_index()
sales_by_category
```

```python
# b) Average customer age by region
avg_age_region = merged_df.groupby("region")["age"].mean().reset_index()
avg_age_region
```

```python
print("Cleaned Data Samples:")
print(customers.head(), "\n")
print("Total Sales by Category:\n", sales_by_category)
print("Average Age by Region:\n", avg_age_region)
```

```python

# Converting into excel file

with pd.ExcelWriter("company_data.xlsx", engine="openpyxl") as writer:
    products.to_excel(writer, sheet_name="Products", index=False)
    customers.to_excel(writer, sheet_name="Customers", index=False)
    sales_merged.to_excel(writer, sheet_name="Sales", index=False)

print("Cleaned data saved successfully to 'company_data.xlsx'")
```
